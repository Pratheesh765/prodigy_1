```python
import pandas as pd

# Load the datasets
pop_data = pd.read_csv("D:/Intership/API_SP.POP.TOTL_DS2_en_csv_v2_633062.csv", skiprows=4)
metadata_indicator = pd.read_csv("D:/Intership/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_633062.csv")
metadata_country = pd.read_csv("D:/Intership/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_633062.csv")

# Display the first few rows of each dataset
print('Population Data:')
print(pop_data.head())
print('\Metadata Indicator:')
print(metadata_indicator.head())
print('\Metadata Country:')
print(metadata_country.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract population data for the year 2020
pop_2020 = pop_data[['Country Name', '2020']].dropna()
pop_2020.columns = ['Country', 'Population_2020']

# Plotting the distribution of population in 2020
plt.figure(figsize=(12, 6))
sns.histplot(pop_2020['Population_2020'], bins=30, kde=True)
plt.title('Distribution of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Extract the income group data
income_group_data = metadata_country[['Country Code', 'IncomeGroup']].dropna()

# Count the number of countries in each income group
income_group_counts = income_group_data['IncomeGroup'].value_counts()

# Plotting the distribution of income groups
plt.figure(figsize=(12, 6))
sns.barplot(x=income_group_counts.index, y=income_group_counts.values)
plt.title('Distribution of Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract the region data
region_data = metadata_country[['Country Code', 'Region']].dropna()

# Count the number of countries in each region
region_counts = region_data['Region'].value_counts()

# Plotting the distribution of regions
plt.figure(figsize=(12, 6))
sns.barplot(x=region_counts.index, y=region_counts.values)
plt.title('Distribution of Regions Among Countries')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Year': [1951, 1961, 1971, 1981, 1991, 2001, 2011],
    'Population': [361088000, 439235000, 548159652, 683329097, 846387888, 1028737436, 1210854977]
}
df = pd.DataFrame(data)
# Calculate the population distribution by decades
df['Decade']= (df['Year']// 10)*10
decade_population = df.groupby('Decade')['Population'].mean().reset_index()

# Plotting the population distribution by decades
plt.figure(figsize=(12, 6))
sns.barplot(data=decade_population, x='Decade', y='Population')
plt.title('Average Population of India by Decades')
plt.xlabel('Decade')
plt.ylabel('Average Population')
plt.show()
```
